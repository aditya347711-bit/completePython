# functional programming 

```python
def add_numbers(a,b):
    print(a+b)
def subtract_numbers(a,b):
    print(a-b)
          
```

```python
subtract_numbers(40,10)
```

```python
add_numbers(50,10)
```

 # object oriented programing 

```python
class calculator:
    def add(self,a,b):
        print(a+b)
    def sub(self,a,b):
        print(a-b)
    def multi(self,a,b):
        print(a*b+2)
```

```python
calc = calculator()
```
