# A Variable name must start with a letter or the underscore character

```python
apple = 23
print(apple)
```

```python
apple_box = 2.3
print(apple_box)
```

```python
_5 = 7
print(_5)
```

```python
_z = 6.3
print(_z)
```

```python
mnp_ = 5 
print(mnp_)
```

# Variable name cannot start with a number

```python
1 == 5 
print(1)
```

```python
1_apple = 213
print(1_apple)
```

```python
apple_1 = 12
print(apple_1)
```

# A Variable name can only contain alpha- numeric characters and underscore (a-z, 0-9,and_)

```python
abc123nmp = 23
print(abc123nmp)
```

```python
abc123456789 = 1
print(abc123456789)
```

```python
#$@ = 25
print(#$@)
```
