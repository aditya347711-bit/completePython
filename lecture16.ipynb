```python
num_students = int(input("enter the number of students"))
course_credits = []
for i in range(num_students):
    credits = int(input("enter the course credit for student:"))
    course_credits.append(credits)
    print(course_credits)

enrollement_status = []
for credits in course_credits:
    if credits>=12:
        enrollement_status.append("full_time")
    else:
        enrollement_status.append("part-time")
for i in range(num_students):
    print(f"students{i+1} : enrollement status - {enrollement_status[i]}")
    
        
    
```
