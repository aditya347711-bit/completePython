# properties of set
# 1) unordered

```python
my_set = {3,1,4,2}
print(my_set)
```

# 2 No indexing/slicing 

```python
# set m aak chiz yad rakhna isme indexing or slicing nahi hota hai
a = {"animal", "ball","cat"}
print(a[0])
```

# 3) unique elements

```python
# unique element mai yad rakhna isme element repeat nhi hota hai
my_list = {2,4,8,6,7,5,4,5,2,8,9,9,6,4,5}
print(my_list)
```

```python
set_in = {"aditya","anil","ankit","aditya","aman","aman"}
print(set_in)
```

# 4) Iterable 

```python
my_set = {12,45,52,36,54}
for element in my_set:
    print(element)
```

# 5) Hasable elements
only hashable objects(objects with a fixed hash value)can be added to set. this means that sets cannot contain mutuable types like lists or dictionary.


```python
# isme tuple hee allowed hai q ki wo immutable hai isme 
my_set = {1,2,3,"a",4,3.4,(10,20)}
print(my_set)
```

```python
# isme list allowed nhi hai
my_set = {1,2,3,4,5,5.4,'b',[21,44,25]}
print(my_set)
```
