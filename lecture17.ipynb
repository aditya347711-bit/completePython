# tuples in python part 1

# creating a tuple

```python
a = (10,20,30)
print(a)
print(type(a))
     
```

```python
tup = ()
print(tup)
print(type(tup))
```

# single element

```python
#single element m tuple use karenge to isme yad rakhna integer type ka data lega isliye (,) lagyange tb tuple show karega
b = (5)
print(b)
print(type(b))
```

```python
# ye right way hai single element likhne ka tuple format mai 
b = (5,)
print(b)
print(type(b))
```

# a tuple with mixed a data type

```python
a = (1,2,3,"aditya","ankit",True)
print(a)
print(type(a))
```

# nested tuples

```python
a = ("mouse", 2.3, (10,20,30))
print(a)
print(type(a))
```

```python
num = ([12,14,16], [14,78,96])
print(num)
print(type(num))
```

 # tuple constructor

```python
empty = tuple()
print(empty)
print(type(empty))
```

# converting list into tuple

```python
list = [1,2,3,4]
number = tuple(list)
print(number)
print(type(number))
```

# converting string into a tuple

```python
string_example = "hello"
print(string_example)
print(type(string_example))
tuple1 = tuple(string_example)
print(tuple1)
print(type(tuple1))
```

# converting dictionary into a tuple

```python
dict_example = {'a':1, 'b':2, 'c':3}
print(dict_example)
print(type(dict_example))
tup = tuple(dict_example)
print(tup)
print(type(tup))
```

# converting a set into a tuple

```python
set_example = {1,2,3,4}
print(set_example)
print(type(set_example))
tup1 = tuple(set_example)
print(tup1)
print(type(tup1))
```
