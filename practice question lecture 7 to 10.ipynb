```python
def grade_distribution(grades):
    for grade in grades:
        if 90<= grade <=100:
            print("you have got A grade")
        elif 80<= grade <=89:
            print("you have got B garde")
        elif 70<= grade <=79:
            print("you have got C garde")
        elif 60<= grade <= 69:
            print("you have got D grade")
        else:
            print("you have got E grade")
    return
```

```python
grades = [85, 92, 88, 74, 63, 95, 70, 58, 82, 67, 99, 73]
grade_distribution(grades)
```

```python
import random
def num_guessing_game():
    secret_number = random.randint(1, 100)
    print(secret_number)
    print("welcome to the number guessing game")
    print("i have choosen the number between 1 to 100. can you guess it ?")
    attempts = 0
    guess = None
    while guess != secret_number:
        guess = int(input("enter your guess"))
        attempts = attempts+1
        if guess < secret_number:
            print("Too low, try again")
        elif guess > secret_number:
            print("Too high, try again")
        else:
            print("congratulation, you have guessed the correct number")
        
```
