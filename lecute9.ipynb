# for loop in python

In python, a for loop is used to iterate over a sequence(such as a list, string, dictonary, or range) and execute a block of code each item in the sequence

```python
number = [1,2,3,4,5]
for i in number:
    print(i)
```

```python
x = "student"
for i in x:
    print(i)
```

```python
for i in range(11):
    print(i)
```

```python
for i in range(10,16):
    print(i)
```

# for loop and if condition 

```python
for i in range(0,21):
    if i%2 ==0:
        print("the number is even",i)
```

```python
for i in range(0,11):
    if i%2 ==0:
        print("the number is even",i)
    else:
        print("the number is odd",i)
```

# Implement a python program to generate the multiplication table a given number using a for loop

```python
number = int(input("enter a number to generate its multiplication table"))
print("the multplication table of",number)
for i in range(1,11):
    print(f"{number} * {i} = {number *i}")
```
