# tuple in python part 2

# indexing

```python
my_tuple = ("apple", "banana", "cherry", "date")
print(my_tuple[0])
print(my_tuple[3])
```

# negative indexing

```python
my_tuple = ("apple", "banana", "cherry", "date")
print(my_tuple[-2])
print(my_tuple[-4])
```

# slicing in tuple

```python
my_tup = ('a','b','c','d','e')
print(my_tup[1:4])
```

```python
my_tup = ('a','b','c','d','e')
print(my_tup[:])
```

```python
my_tup = ('a','b','c','d','e')
print(my_tup[::2])
```

# slicing - negative indexing


```python
tup2 = ('apple','banana','cherry','date')
print(tup2[:-3])
print(tup2[-3:])
```

# reversing the tuple

```python
a = (45,21,31,14,54,65,12)
print(a[::-1])
```

# Duplicate are allowed in tuple

```python
a = (20,54,67,55,23,47)
print(a)
```

# Method of tuples

# count method

```python
a = (12,14,18,23,14,45,18)
print(a.count(18))
```

```python
b = ("aditya","anil","ankit","jay","aditya")
print(b.count("aditya"))
```

# index

```python
b = ('aditya','ankit','aman', 'anil')
print(b.index('aditya'))
print(b.index('aman'))
```
