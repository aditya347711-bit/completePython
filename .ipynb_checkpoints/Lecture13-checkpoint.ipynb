# length - method

```python
s = "theiscale"
print(len(s))
```

```python
s = " "
print(len(s))
```

```python
s = "  "
print(len(s))
```

```python
s = "hello\nworld"
print(s)
print(len(s))
```

```python
s = "hello@world"
print(s)
print(len(s))
```

```python
s = """hello
world"""
print(s)
print(len(s))
```

```python
a = "data science is best carrier path for the student"
print(len(a))
```
