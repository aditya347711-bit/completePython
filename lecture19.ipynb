# once a tuple is created, you cannot change its values. tuple are unchangable, or immutable as it also is called.

# convert tuple into list

```python
x = ("data","python","science","swati")
print(x)
print(type(x))

y = list(x)
print(y)
print(type(y))
```

# add items

```python
a = (25,2,27,28,29)
y = list(a)
print(y)
print(type(y))

y.append(30)
print("the new list is")
print(y)

x = tuple(y)
print("the new item added on tuple is ")
print(x)
```

# Add tuple to a tuple 

```python
tup1 = ("python","data","science",'aditya')
y = ("code",)
tup1+=y
print(tup1)
```

# Remove items

```python
fruits = ("apple","banana","cherry","date","fig")
fruit_list = list(fruits)
print(fruit_list)

fruit_list.remove("cherry")
print(fruit_list)

fruit_tup = tuple(fruit_list)
print(fruit_tup)
```

```python
x = ("apple","banan","cherry","date","fig")
y = list(x)
print(y)
```

# loop through a tuple

```python
thistuple = ("a","b","c")
for x in thistuple:
    print(x)
```

# join two tuples

```python
tuple1 = ("a","b","c")
tuple2 = (1,2,3)

tuple3 = tuple1+tuple2
print(tuple3)
```

# multiply tuples

```python
fruit = ("apple","banana","cherry")
mytup = fruit*3
print(mytup) 
```
