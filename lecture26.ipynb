# change values

```python
a = {"name": "aditya",
     "age": 21,
     "country": "india"}
print(a)
     
```

```python
a["country"] = "bharat"
print(a)
```

```python
a["age"] = 22
print(a)
```

```python
a["name"] = "ankit"
print(a)
```

# Adding items

```python
person = {"name": "aditya",
          'age':30}
print(person)
          
```

```python
person["city"] = "Dhanbad"
person["email id"] = "aditya347711@gmail.com"
print(person)
```

# update method

```python
person = {"name": "aditya",
          "age": 21,
          "country": "india",
          "gender": "male"}
print(person)
          
          
```

```python
person.update({"phone number":9031773602})
print(person)
```

```python
person.update({"mail id":"aditya347711@gmail.com", "adhar number": 9142807736})
print(person)
```

# Remove items

```python
a = {"name": "aditya","age":21,"country":"india","gender":"male"}
print(a)
```

```python
a.pop("age")
print(a)
```

# del method 

```python
a = {"name": "aditya","age":21,"country":"india","gender":"male"}
```

```python
del a["country"],a["gender"]
print(a)
```
