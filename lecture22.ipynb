# intersection method

```python
set1 = {25,56,32}
set2 = {25,89,27}

inter = set1.intersection(set2)
print(inter)
```

# practice question

```python
#define the two sets 
students_passed_maths = {101,102,103,104,105}
students_passed_science = {103,104,105,106,107}
```

```python
both_passed = students_passed_maths.intersection(students_passed_science)
print("student id is those students who have passed in maths and science",both_passed)
print("total studentspassed in both maths and science",len(both_passed))
```

```python
only_maths = students_passed_maths.difference(students_passed_science)
print("those who hase passed in math only",only_maths)
```

```python
only_science = students_passed_science.difference(students_passed_maths)
print("those who hase passed in science only",only_science)
```

```python
sub = students_passed_maths.issubset(students_passed_science)
print("is students passed in maths is a subset of the students passed in science",sub)
```
