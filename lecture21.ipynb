# true and 1 are considered as same value

```python
# yad rakhna set m true ko 1 se consider karte hai
a = {'animal',"ball",1,20,52,5.4,56,True}
print(a)
```

```python
# yad rakhna set m False ko 0 se consider karte hai
a = {'animal',"ball",1,20,52,5.4,56,True,0,False}
print(a)
```

```python
#agar isme 1 nhi likhe to ye true hi output dega
a = {'animal',"ball",20,52,5.4,56,True}
print(a)
```

```python
#agar isme 0 nhi likhe to ye false hi output dega
a = {'animal',"ball",20,52,5.4,56,False}
print(a)
```

# get the length of the set

```python
# isme 10 isliye aa rha q ki 23 ,23 repeat hai 2 bar to set m ak hi count karte hai
a = {"animal","bat","cat",12,54,6,45,23,87,23,56}
print(len(a))
```

# Add items

# once a set is created, you cnnot change the items but you can add the new items

```python
a = {"a1","b1","c1",1,2,3}
a.add("d1")
print(a)
```

# update method

```python
x = {"data","python","code"}
y = {10,20,30,40,50,60,70}

x.update(y)
print(x)


```

# remove items from the set

```python
n = {70,40,10,"code","data","science",42,12,36,77,23,5.2,6.6}
n.remove("code")
print(n)
```

# pop method

```python
# ye random ak no remove kr dega 
a = {15,23,45,75,96,521,89}
c = a.pop()
print(a)
print(c)
```

```python
x = {'code','apple','aditya',23,56,25,45,9.4}
y = x.pop
print(y)
print(x)
```

# clear method

```python
x = {'code','apple','aditya',23,56,25,45,9.4}
x.clear()
print(x)
```
