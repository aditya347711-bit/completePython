# creating simple dictionary

```python
person = {
    'name' : "aditya",
    'age' : 21,
    'city' : 'BANGALORE'
}
```

```python
print(person)
print(type(person))
```

# Dictionary with different data types

```python
mixed_dict = {
    'name' : 'aditya',
    'age' : 21,
    'is_student' : True,
    'courses' : ['maths','physics','chemistry'],
    'address' : {
        'street' : "digwadih no10",
        'city' : "Dhanbad",
        'code' : 828110}
}
    
```

```python
print(mixed_dict)
print(type(mixed_dict))
```

# Empty dectionary

```python
empty = {}
print(empty)
print(type(empty))
```

# Dict constructor

```python
person = dict(name = 'aditya', age = 21, city = 'jharkhand')
print(person)
print(type(person))
```

# Dictionary comprehension

```python
square = {x: x*x for x in range(6)}
print(square)
```
